```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('housing.csv', sep=',', encoding='latin-1')
df.head() 
```

```python
df ['number of bedrooms relative to all rooms'] = df['total_bedrooms']/df['total_rooms']
```

```python
df.head()
```

```python
df.shape
```

```python
df_non_numeric = df.select_dtypes(exclude=[np.number])
non_numeric_cols = df_non_numeric.columns.values
print(non_numeric_cols)
```

```python
del df['ocean_proximity']
```

```python
Q1 = df.quantile(q=.25)
Q3 = df.quantile(q=.75)
IQR = df.apply(stats.iqr)
data_clean = df[~((df < (Q1-1.5*IQR)) | (df > (Q3+1.5*IQR))).any(axis=1)]
data_clean.shape
```

```python
data_clean.drop_duplicates ()
```

```python
(data_clean.isnull().sum()/len(data_clean))*100
```

```python
data_clean=data_clean[data_clean.total_bedrooms>0]
```

```python
data_clean.shape
```

```python
# Для анализа зависимости будем использовать housing_median_age (средний возраст дома в квартале), 
# population (общее количество людей, проживающих в квартале) и number of bedrooms relative to all rooms 
# (общее количество домохозяйств, группа людей, проживающих в пределах жилой единицы, для квартала)
# к total_rooms (общее количество комнат в блоке), total_bedrooms ( общее количество спален в блоке)
# и median_income (средний доход домохозяйств в пределах многоквартирного дома (измеряется в десятках тысяч долларов США))
```

```python
select = ['housing_median_age', 'median_income']
select_df = data_clean[select]
select_df
```

```python
Nc = range(1,20)
kmeans = [KMeans(n_clusters=i,n_init=10) for i in Nc]
score = [kmeans[i].fit(select_df).score(select_df) for i in range(len(kmeans))]
```

```python
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.show()
```

```python
# Судя по "Локтю" нам нужно 4 кластера
```

```python
kmeans=KMeans(n_clusters=5, n_init = 10)
select_df.loc[:,('KMean')] = kmeans.fit_predict(select_df)
```

```python
select_df.head()
```

```python
for i in range(0, 5):
    draw_df = select_df[select_df['KMean'] == i]  
    plt.scatter(draw_df['housing_median_age'], draw_df['median_income'],  alpha=0.01)
```
