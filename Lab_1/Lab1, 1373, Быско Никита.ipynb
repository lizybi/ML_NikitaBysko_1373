a) Информатика, социальная наука; Перепись населения Калифорнии 1990 года; реальные
b) 1. longitude: показатель того, насколько далеко на западе находится дом; более высокое значение находится дальше на запад
   2. latitude: мера того, насколько далеко на севере находится дом; более высокое значение находится дальше на север
   3. housingMedianAge: средний возраст дома в квартале; меньшее число означает более новое здание
   4. totalRooms: общее количество комнат в блоке.
   5. totalBedrooms: общее количество спален в блоке.
   6. population: общее количество людей, проживающих в квартале.    7. households: общее количество домохозяйств, группа людей, проживающих в пределах жилой единицы, для квартала.
   8. medianIncome: средний доход домохозяйств в пределах многоквартирного дома (измеряется в десятках тысяч долларов США).
   9. medianHouseValue: медианная стоимость дома для домохозяйств внутри квартала (измеряется в долларах США).
   10. oceanProximity: расположение дома относительно океана/моря
c) нет
